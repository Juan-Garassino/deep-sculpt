```python
from deepSculpt.components.edges import add_edge
from deepSculpt.components.planes import add_plane
from deepSculpt.plotter import Plotter

import numpy as np

from deepSculpt.params import (
    VOID_DIM,
    N_SAMPLES,
    BUCKET_NAME,
    BUCKET_TRAIN_DATA_PATH,
    MODEL_BASE_PATH,
    N_EDGE_ELEMENTS,
    N_PLANE_ELEMENTS,
    N_VOLUME_ELEMENTS,
    COLOR_EDGES,
    COLOR_PLANES,
    COLOR_VOLUMES,
    ELEMENT_EDGE_MIN,
    ELEMENT_EDGE_MAX,
    ELEMENT_GRID_MIN,
    ELEMENT_GRID_MAX,
    ELEMENT_PLANE_MIN,
    ELEMENT_PLANE_MAX,
    ELEMENT_VOLUME_MIN,
    ELEMENT_VOLUME_MAX,
    STEP,
    VERBOSE
)
```

```python
void = np.zeros((VOID_DIM, VOID_DIM, VOID_DIM))

color_void = np.empty(void.shape, dtype=object)

edge = add_edge(void, color_void, ELEMENT_EDGE_MIN, ELEMENT_EDGE_MAX, STEP, VERBOSE)

type(edge)
```

```python
void = np.zeros((VOID_DIM, VOID_DIM, VOID_DIM))

color_void = np.empty(void.shape, dtype=object)

plane = add_plane(void, color_void, ELEMENT_PLANE_MIN, ELEMENT_PLANE_MAX, STEP, VERBOSE)

type(plane)

```
