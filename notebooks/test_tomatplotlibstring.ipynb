```python
%load_ext autoreload
%autoreload 2

COLORS = dict(edges="dimgrey",
              planes="snow",
              volumes=["crimson", "turquoise", "gold"])
```

```python
from deepSculpt.collector.collector import Collector
from deepSculpt.manager.tools.plotter import Plotter
from deepSculpt.manager.manager import Manager
import os
import numpy as np

collector = Collector(
    void_dim=int(os.environ.get("VOID_DIM")),
    edge_elements=(0, 0.3, 0.5),
    plane_elements=(0, 0.3, 0.5),
    volume_elements=(2, 0.3, 0.5),
    step=None,
    directory='.',
    chunk_size=50,
    n_chunks=1,
    grid=1,
    #colors=COLORS
)

# Creates the data
volumes_void, materials_void = collector.create_collection()
```

```python
size = 16

arr = np.random.randint(0, 2, size=(size, size, size))

arr_color = Manager().convert_to_matplotlib_colors(
    np.random.randint(0, 256, size=(size, size, size, 3)))

plot = Plotter().plot_sculpture(
    volumes=arr,
    materials=arr_color,
    directory='.',
    raster_picture=True,
    vector_picture=False,
    volumes_array=False,
    materials_array=False,
    hide_axis=True,
)
```
